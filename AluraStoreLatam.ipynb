### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

```

#1. Análisis de facturación



```python
facturacion_t1 = tienda['Precio'].sum()
facturacion_t2 = tienda2['Precio'].sum()
facturacion_t3 = tienda3['Precio'].sum()
facturacion_t4 = tienda4['Precio'].sum()

print(f"Ingreso Total Tienda 1: ${facturacion_t1:,.2f}")
print(f"Ingreso Total Tienda 2: ${facturacion_t2:,.2f}")
print(f"Ingreso Total Tienda 3: ${facturacion_t3:,.2f}")
print(f"Ingreso Total Tienda 4: ${facturacion_t4:,.2f}")
```

```python
import matplotlib.pyplot as plt

# Preparación de los datos para el gráfico
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [facturacion_t1, facturacion_t2, facturacion_t3, facturacion_t4]

# Configuración del gráfico
plt.figure(figsize=(10, 5))
plt.bar(tiendas, ingresos, color=['skyblue', 'lightgreen', 'salmon', 'orchid'])

# Diseño de detalles estéticos
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tiendas Alura')
plt.ylabel('Facturación ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Publicación del gráfico
plt.show()
```

# 7. Analisis de corelación entre productos max, min y calificación de esos producto.

#### Correlación productos más vedidos por tienda y calificación promedio de esos productos.

```python
producto_max_t1 = tienda['Producto'].value_counts().max()
producto_max_t2 = tienda2['Producto'].value_counts().max()
producto_max_t3 = tienda3['Producto'].value_counts().max()
producto_max_t4 = tienda4['Producto'].value_counts().max()

nombre_max_t1 = tienda['Producto'].value_counts().idxmax()
nombre_max_t2 = tienda2['Producto'].value_counts().idxmax()
nombre_max_t3 = tienda3['Producto'].value_counts().idxmax()
nombre_max_t4 = tienda4['Producto'].value_counts().idxmax()

calificaciones_top_t1 = tienda[tienda['Producto'] == nombre_max_t1]['Calificación']
calificaciones_top_t2 = tienda2[tienda2['Producto'] == nombre_max_t2]['Calificación']
calificaciones_top_t3 = tienda3[tienda3['Producto'] == nombre_max_t3]['Calificación']
calificaciones_top_t4 = tienda4[tienda4['Producto'] == nombre_max_t4]['Calificación']


promedio_top_t1 = calificaciones_top_t1.mean()
promedio_top_t2 = calificaciones_top_t2.mean()
promedio_top_t3 = calificaciones_top_t3.mean()
promedio_top_t4 = calificaciones_top_t4.mean()
```

```python
print(f"Tienda 1: El producto '{nombre_max_t1}' vendió {producto_max_t1} unidades y su calificación promedio es {promedio_top_t1:.2f}")
print(f"Tienda 2: El producto '{nombre_max_t2}' vendió {producto_max_t2} unidades y su calificación promedio es {promedio_top_t2:.2f}")
print(f"Tienda 3: El producto '{nombre_max_t3}' vendió {producto_max_t3} unidades y su calificación promedio es {promedio_top_t3:.2f}")
print(f"Tienda 4: El producto '{nombre_max_t4}' vendió {producto_max_t4} unidades y su calificación promedio es {promedio_top_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Datos que calculaste previamente
ventas_eje_x = [producto_max_t1, producto_max_t2, producto_max_t3, producto_max_t4]
calificaciones_eje_y = [promedio_top_t1, promedio_top_t2, promedio_top_t3, promedio_top_t4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['skyblue', 'lightgreen', 'salmon', 'orchid']

# 2. Configuración del gráfico
plt.figure(figsize=(10, 6))

# Dibujamos los puntos
# s=500 aumenta el tamaño del punto para que sea más visible
plt.scatter(ventas_eje_x, calificaciones_eje_y, s=500, c=colores, edgecolors='black', alpha=0.8)

# 3. Añadir etiquetas a cada punto para identificar la tienda
for i, txt in enumerate(nombres_tiendas):
    plt.annotate(txt, (ventas_eje_x[i], calificaciones_eje_y[i]),
                 textcoords="offset points", xytext=(0,15), ha='center', fontsize=12, fontweight='bold')

# 4. Personalización estética
plt.title('Correlación: Unidades Vendidas vs. Calificación Promedio (Top Productos)', fontsize=16)
plt.xlabel('Cantidad de Unidades Vendidas', fontsize=13)
plt.ylabel('Calificación Promedio (1-5)', fontsize=13)
plt.ylim(0, 5.5) # Escala de calificación
plt.grid(True, linestyle='--', alpha=0.6)

# 5. Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Datos
productos = [nombre_max_t1, nombre_max_t2, nombre_max_t3, nombre_max_t4]
ventas = [producto_max_t1, producto_max_t2, producto_max_t3, producto_max_t4]
calificaciones = [promedio_top_t1, promedio_top_t2, promedio_top_t3, promedio_top_t4]

fig, ax1 = plt.subplots(figsize=(12, 7))

# --- 1. BARRAS: UNIDADES VENDIDAS ---
color_ventas = '#3498db'
# Dibujo de las barras
barras = ax1.bar(productos, ventas, color=color_ventas, alpha=0.7, label='Unidades Vendidas', width=0.5)

ax1.set_xlabel('Productos Estrella por Tienda', fontsize=12, fontweight='bold')
ax1.set_ylabel('Unidades Vendidas', color=color_ventas, fontsize=14, fontweight='bold')
ax1.tick_params(axis='y', labelcolor=color_ventas)

# Etiquetas de valor sobre las BARRAS
for barra in barras:
    yval = barra.get_height()
    ax1.text(barra.get_x() + barra.get_width()/2, yval + 1, f'{int(yval)}',
             ha='center', va='bottom', color=color_ventas, fontweight='bold', fontsize=12)

# --- 2. LÍNEA: CALIFICACIÓN ---
ax2 = ax1.twinx()
color_calif = '#e74c3c'
# Dibujo de la línea con puntos
ax2.plot(productos, calificaciones, color=color_calif, marker='o', markersize=10,
         linewidth=3, label='Calificación Promedio', zorder=5)

ax2.set_ylabel('Calificación Promedio', color=color_calif, fontsize=14, fontweight='bold')
ax2.tick_params(axis='y', labelcolor=color_calif)
ax2.set_ylim(0, 5.5)

# Añadir etiquetas de valor sobre los PUNTOS de la línea
for i, v in enumerate(calificaciones):
    ax2.text(i, v + 0.15, f'{v:.2f}', ha='center', color=color_calif, fontweight='bold', fontsize=12)

# Título y ajustes finales
plt.title('Rendimiento de Productos Estrella: Volumen vs. Calidad', fontsize=16, pad=20)
ax1.grid(True, axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
